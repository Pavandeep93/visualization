```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
np.random.seed(613)
x = np.arange(50)
y =np.random.randint(0,100,50)

fig, ax = plt.subplots(figsize = (5,3))
#ax.scatter(x,y)
#ax.bar(x,y)
#ax.plot(x,y)
ax.hist(y) # only using y axis 
```

```python
font1 = {'family': 'sans-serif','color':'blue', 'size':20}
font2 = {'family': 'monospace','color':'green','size':14}



fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x,y)

ax.set_title('Total growth over time', fontdict=font1, loc = 'left')
ax.set_ylabel('Total growth', fontdict = font2)
ax.set_xlabel('Years since start', fontdict = font2)
```

```python
fig, ax = plt.subplots(figsize=(5,3))
ax.scatter(x,y,
           marker = '*',
           color = 'indigo')
```

```python
fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x,y,
        marker = '*',
        color = 'indigo',
        linestyle = '--',
        linewidth = 2,
        markersize = 12,
        markeredgecolor = '#fa9359',
        markerfacecolor = '#000000')

ax.grid(color = 'green',          ## axis = 'x' or axis = 'y' for vertical or horizontal lines only
        linestyle = 'dotted',
        linewidth = 3)
```

## Activity 

```python
height = [3, 12, 5, 18, 45]
bars = ('A', 'B', 'C', 'D', 'E')

df = pd.DataFrame({'height': height,
                   'bars': bars})
```

```python
color = (0.2, # redness
         0.4, # greenness
         0.2, # blueness
         0.6 # transparency
         ) 

# Create bars
fig, ax = plt.subplots()
ax.bar(df['bars'], df['height'],
       color=color)

# Show graph
plt.show()
```

```python
color = ['lightblue', 'blue', 'purple', 'red', 'black']

# Create bars
fig, ax = plt.subplots()
ax.bar(df['bars'], df['height'],
       color=color)

# Show graph
plt.show()
```
