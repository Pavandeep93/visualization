```python
# Modifying our legend

np.random.seed(613)
x = np.arange(50)
y1 = np.random.randint(0,100,50)
y2 = np.random.randint(0,100,50)

fig, ax = plt.subplots(figsize = (5,3))
ax.plot(x,y1, label = "Person 1")
ax.plot(x,y2, label = "Person 2")
ax.legend(loc = 'lower right',
          frameon = True,
          fontsize = 12,
          ncol = 2,
          shadow = True)
```

```python
# Moving the legend outside plot area

fig, ax = plt.subplots(figsize = (5,3))
ax.plot(x,y1, label = "Person 1") # fist line for plot
ax.plot(x,y2, label = "Person 2")  # second line for plot
ax.legend(loc = 'upper left',
          bbox_to_anchor = (1,1)) # legend moved to upper left corner
```

```python
# Text and Annotations
# Annotations with text() functions

fig, ax = plt.subplots(figsize = (5,3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc = 'lower right')

ax.text(10,95, "This value is important")
```

```python
# Modifying text annotattions
fig, ax = plt.subplots(figsize = (5,3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc = 'lower right')

ax.text(10,95, "This value is important!",
        ha = 'center',  # alignment of text
        color = 'red',
        size = 20)
```

```python
# Positionning text annotations
fig, ax = plt.subplots()
ax.axis([0, 10, 0, 10])  # empty set of axes

ax.text(1,5, "Data: (1,5)", transform = ax.transData)
ax.text(0.5, 0.1, "Axes: (0.5, 0.1)", transform = ax.transAxes)
ax.text(0.2, 0.2, "Figure: (0.2, 0.2)", transform = fig.transSubfigure)
```

```python
fig, ax = plt.subplots(figsize = (5,3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc = 'lower right')

ax.annotate("This is important!", xy = (10,95), xytext= (20,94),
            arrowprops= dict(facecolor = 'black', shrink = 0.05)) # shrink changes the distance between text
```

```python
# Modifying the arrow
fig, ax = plt.subplots(figsize = (5,3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc = 'lower right')

ax.annotate("This is important!", xy = (10,95), xytext= (20,94),
            arrowprops= dict(arrowstyle = "wedge", color = "hotpink"))
```

```python
# Axis labels, Removing tick marks and labels
fig, ax = plt.subplots(figsize = (5,3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc = 'lower right')

ax.yaxis.set_major_locator(plt.NullLocator())
ax.xaxis.set_major_formatter(plt.NullFormatter())
```

```python
# Limting tick marks

fig, ax = plt.subplots(figsize = (5,3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc = 'lower right')

ax.yaxis.set_major_locator(plt.NullLocator())
ax.xaxis.set_major_formatter(plt.NullFormatter())

ax.xaxis.set_major_locator(plt.MaxNLocator(3))
```

```python
# Tick mark intervals(mulitply by 5)

fig, ax = plt.subplots(figsize = (5,3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc = 'lower right')

ax.xaxis.set_major_locator(plt.MultipleLocator(5))

plt.xticks(rotation = 45, ha = 'right', fontsize= 12, font = "serif") # change the ticks to rotate 45 and adding customization to font style
```

# Activity : Customizing plots

```python
fig,(ax1, ax2) = plt.subplots(ncols=2,
                              nrows=1,
                              figsize=(7,3))

color = (0.2, # redness
         0.4, # greenness
         0.2, # blueness
         0.6 # transparency
         )

ax1.scatter(x1,y1,
            marker= "*",
            color= "orange")
ax2.bar(x2,y2,
        color= color)

font1 = {'family': 'sans-serif','color':'green', 'size':20}
font2 = {'family': 'monospace','color':'red','size':12}

ax1.set_title('Activity', fontdict=font1, loc = 'left')
ax2.set_title('Activity 2', fontdict=font1, loc = 'left')

ax1.set_ylabel('test 1', fontdict = font2)
ax2.set_ylabel('test 2', fontdict = font2)

ax1.grid(color = 'green',          
        linestyle = 'dotted',
        linewidth = 1)


plt.xticks(rotation = 45, ha = 'right', fontsize= 12, font = "serif")
```

```python
# Subplots without a grid arrangement 

fig, someaxes = plt.subplot_mosaic([['ax1', 'ax3',],
                                     ['ax2','ax3']],
                                    figsize = (7,4))
someaxes["ax1"].scatter(x1,y1)
someaxes["ax2"].bar(x2,y2)
someaxes["ax3"].plot(x1,y1)
```

```python
# Layouts 
fig, someaxes = plt.subplot_mosaic([['ax1', 'ax3',],
                                     ['ax2','ax3']],
                                    figsize = (7,4),
                                    layout = "constrained") # layout adjustment to set labels
someaxes["ax1"].scatter(x1,y1)
someaxes["ax2"].bar(x2,y2)
someaxes["ax3"].plot(x1,y1)

someaxes["ax1"].set_xlabel('A Big Label', fontsize=18)
someaxes["ax2"].set_xlabel('Another Label', fontsize =18)
someaxes["ax3"].set_xlabel('Label 2: 2 fast 2 Furious' , fontsize = 18)
```

```python
# Multiple viz on one axes

x = np.array(["Luffy", "Zoro", "Nami", "Usopp", "Sanji"])
y1 = np.array ([110,180,240,99,220])
y2 = np.array([170,100,90,120,50])

fig, ax = plt.subplots(figsize= (7,3))
ax.bar(x,y1,
       color = "indigo")
ax.plot(x,y2,
        color = "red")
```

```python
y2_sd = np.std(y2)

fig, ax = plt.subplots(figsize=(7,3))
ax.plot(x,y2, color = 'red')

ax.errorbar(x,y2,
            yerr=y2_sd,
            fmt = "none",
            ecolor = "indigo",
            elinewidth= 4,
            capsize= 6,
            capthick= 4,
            errorevery= 2)
```
